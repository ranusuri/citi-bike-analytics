```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import requests
import glob
import os
```

```python
# combine all 2022 monthly data files into single csv file
# merging the files
#files_joined = os.path.join('C:\\Users\\DELL\\OneDrive\\Desktop\\CSV_files', "salesdata*.csv")

# Return a list of all joined files
list_files = glob.glob("data/*.csv")

print("** Merging multiple csv files into a single pandas dataframe **")
# Merge files by joining all files
df = pd.concat(map(pd.read_csv, list_files), ignore_index=True)

print(df)
```

```python
df['date'] = pd.to_datetime(df['starttime'])

# add a column for Year
df['year'] = df['date'].dt.year
# add a column for Month
df['month'] = df['date'].dt.month
# add a column for Day
df['day'] = df['date'].dt.day
# add a column for Day of the Year
df['dayofyear'] = df['date'].dt.dayofyear
# add a column for Day of the Year
df['hour'] = df['date'].dt.hour

#age
df['age'] = df['year'] - df['birth year']

df.head()
```
